```python
# 2. Visualizing Data Distributions for Generative AI
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data (two groups with different means & spreads)
data_group1 = np.random.normal(loc=50, scale=10, size=500)    # Group 1
data_group2 = np.random.normal(loc=200, scale=15, size=500)   # Group 2

# --- Visualization ---
plt.figure(figsize=(10, 6))

# Plot overlapping histograms with transparency
plt.hist(data_group1, bins=30, alpha=0.6, color="royalblue", edgecolor="black", label="Group 1")
plt.hist(data_group2, bins=30, alpha=0.6, color="seagreen", edgecolor="black", label="Group 2")

# Styling
plt.title("Data Distributions for Generative AI", fontsize=14, fontweight="bold")
plt.xlabel("Data Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

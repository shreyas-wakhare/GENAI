```python
# 9. Simulating Diffusion Denoising
# ---------------------------------
# Example 9: Simulating Diffusion Denoising
# ---------------------------------

import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a random noisy image (28x28)
image = np.random.rand(28, 28)

# 2. Display initial noisy image
plt.imshow(image, cmap='gray')
plt.title("Step 0: Noise")
plt.axis('off')
plt.show()

# 3. Simulate denoising over 3 steps
for step in range(1, 4):
    image = image * 0.9  # reduce noise gradually
    plt.imshow(image, cmap='gray')
    plt.title(f"Step {step}: Denoising")
    plt.axis('off')
    plt.show()
```

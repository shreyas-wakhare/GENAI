```python
#1 preprocessing

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data (10 samples, 5 features)
data = np.random.randint(0, 255, (10, 5))

# Initialize MinMaxScaler (scales between 0 and 1 by default)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Display original and scaled data
print("ðŸ”¹ Original Data:\n", data)
print("\nðŸ”¹ Scaled Data (0-1 range):\n", scaled_data)

# --- Visualization ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot original data
axes[0].imshow(data, cmap="viridis", aspect="auto")
axes[0].set_title("Original Data")
axes[0].set_xlabel("Features")
axes[0].set_ylabel("Samples")

# Plot scaled data
axes[1].imshow(scaled_data, cmap="viridis", aspect="auto")
axes[1].set_title("Scaled Data (0 - 1)")
axes[1].set_xlabel("Features")
axes[1].set_ylabel("Samples")

plt.suptitle("Min-Max Scaling Transformation", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

```
